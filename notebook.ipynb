```python
data = {'true': y_test_b, 'pred': y_pred_b}
results = pd.DataFrame(data)

results.plot(figsize=(20,10), kind='scatter', x='true', y='pred')
```

# 7. Tuning

## 7.1. Best Parameter

```python
learning_rate = [0.001, 0.005, 0.01, 0.002, 0.05, 0.1, 0.2, 0.3]

n_estimators = [int(x) for x in np.linspace(start = 100, stop = 2000, num = 20)]

subsample = [0.8, 0.9, 1.0]

min_samples_split = [2, 5, 10, 20, 40]

min_samples_leaf = [1, 2, 4, 10, 20]

max_features = [None, 'sqrt']

max_depth = [10,20,50,100,150,200,300,500]
max_depth.append(None)
```

```python
import pprint as pp

# Create the random grid
grad_grid = {'n_estimators': n_estimators,
             'learning_rate': learning_rate,
             'subsample': subsample,
             'min_samples_split': min_samples_split,
             'min_samples_leaf': min_samples_leaf,
             'max_features': max_features,
             'max_depth': max_depth}
pp.pprint(grad_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV

gb = GradientBoostingRegressor()

gb_random = RandomizedSearchCV(estimator = gb, param_distributions = grad_grid, n_iter = 50, cv = 3, verbose=2, random_state=42, n_jobs = -1)

# Fit the random search model
gb_random.fit(X_train, y_train)

performances = pd.DataFrame.from_dict(gb_random.cv_results_)
```

```python
sorted_performances = performances.sort_values(by='rank_test_score')
sorted_performances.head()
```
